# Problem Statement:
- For pairs of brands in the same year (e.g. apple/samsung/2020 and samsung/apple/2020) 
      - if custom1 = custom3 and custom2 = custom4 : then keep only one pair
- For pairs of brands in the same year 
      - if custom1 != custom3 OR custom2 != custom4 : then keep both pairs
- For brands that do not have pairs in the same year : keep those rows as well"						
						
						
						
						
						


```python
import pandas as pd

d = {
    'BRAND1': ['apple', 'samsung', 'apple', 'samsung', 'google', 'oneplus'],
    'BRAND2': ['samsung', 'apple', 'samsung', 'apple', '', 'nothing'],
    'YEAR': [2020, 2020, 2021, 2021, 2020, 2020],
    'CUSTOM1': [1, 1, 1, 5, 5, 5],
    'CUSTOM2': [2, 2, 2, 3, 9, 9],
    'CUSTOM3': [1.0, 1.0, 5.0, 1.0, float('nan'), 6.0],
    'CUSTOM4': [2.0, 2.0, 3.0, 2.0, float('nan'), 3.0]
}
df = pd.DataFrame(d)
df
```

```python
result = df.drop_duplicates(subset=["YEAR", "CUSTOM1","CUSTOM2", "CUSTOM3", "CUSTOM4"])
result
```
