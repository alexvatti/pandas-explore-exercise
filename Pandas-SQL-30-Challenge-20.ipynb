```python
result_df = pd.DataFrame(columns=df.columns)
result_df
```

```python
result = df.groupby("COUNTRY")

for country, row in result:
    l = len(row)
    #print(country, l)
    
    if l == 1:
        #print(row)
        # Append row to result_df
        result_df = pd.concat([result_df, row], ignore_index=True)
        
    elif l == 2:
        #print(row.iloc[0])
        #print(row.iloc[1])
        # Append both rows to result_df
        result_df = pd.concat([result_df, row], ignore_index=True)
        
    elif l > 2 and l % 2 == 0:
        row.sort_values("AGE", inplace=True)
        #print(row.iloc[l // 2 - 1])
        #print(row.iloc[l // 2])
        # Append the two median rows to result_df
        result_df = pd.concat([result_df, row.iloc[[l // 2 - 1, l // 2]]], ignore_index=True)
        
    else:
        row.sort_values("AGE", inplace=True)
        #print(row.iloc[l // 2])
        # Append the median row to result_df
        result_df = pd.concat([result_df, row.iloc[[l // 2]]], ignore_index=True)

# Final DataFrame with the appended rows
#print("\nAppended Rows DataFrame:")
print(result_df[["COUNTRY","AGE"]].sort_values(["COUNTRY","AGE"]))
        
```
