```python
import pandas as pd

# Step 1: Create a list of tuples representing the employee and their manager
data = [
    (1, 'Sundar', None),
    (2, 'Kent', 1),
    (3, 'Ruth', 1),
    (4, 'Alison', 1),
    (5, 'Clay', 2),
    (6, 'Ana', 2),
    (7, 'Philipp', 3),
    (8, 'Prabhakar', 4),
    (9, 'Hiroshi', 4),
    (10, 'Jeff', 4),
    (11, 'Thomas', 1),
    (12, 'John', 15),
    (13, 'Susan', 15),
    (14, 'Lorraine', 15),
    (15, 'Larry', 1)
]

# Step 2: Create a DataFrame from the list
df = pd.DataFrame(data, columns=['id', 'name', 'manager'])

# Display the DataFrame
print(df)
```

```python
result = df.groupby(["manager"]).count().reset_index()
result_dict = df.set_index('id')['name'].to_dict()
result["manager"]=result["manager"].astype('int')
result["manager"]=result["manager"].map(result_dict)
result.rename(columns={"id":"NO_OF_EMPLOYEES" ,"manager":"MANAGER"},inplace=True)
result.drop(columns="name",inplace=True)
result.sort_values("NO_OF_EMPLOYEES",ascending=False).reset_index()
```
