```python
import pandas as pd

# Data for INPUT
data = {
    "PRODUCT_ID": [
        'AP755', 'AP755', 'AP755', 'AP755', 'AP755', 'AP755', 'AP755', 
        'XQ802', 'XQ802', 'XQ802', 'XQ802', 'XQ802', 'XQ802', 'XQ802'
    ],
    "DAY_INDICATOR": [
        '1010101', '1010101', '1010101', '1010101', '1010101', '1010101', '1010101', 
        '1000110', '1000110', '1000110', '1000110', '1000110', '1000110', '1000110'
    ],
    "DATES": [
        '04-03-2024', '05-03-2024', '06-03-2024', '07-03-2024', '08-03-2024', '09-03-2024', '10-03-2024',
        '04-03-2024', '05-03-2024', '06-03-2024', '07-03-2024', '08-03-2024', '09-03-2024', '10-03-2024'
    ]
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
weekdays = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Function to convert binary string to a dictionary of days with 1 (available) or 0 (not available)
def day_mapper(day_indicator):
    return {weekdays[i]: int(day_indicator[i]) for i in range(7)}

# Apply the function to each row in the DataFrame
df['DAY_MAPPING'] = df['DAY_INDICATOR'].apply(day_mapper)
df
```

```python
df["DATES"]=pd.to_datetime(df["DATES"])
df
```

```python
df["DAY"]=df["DATES"].apply(lambda x: x.dayofweek + 1)
weekdays = {1:"Mon", 2:"Tue", 3:"Wed", 4:"Thu", 5:"Fri", 6:"Sat", 7:"Sun"}
df["DAY"] = df["DAY"].map(weekdays)
```

```python
df
```

```python
df["Status"] = df.apply(lambda x: 1 if x["DAY_MAPPING"].get(x["DAY"]) == 1 else 0, axis=1)
df
```
