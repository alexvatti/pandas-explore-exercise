```python
import pandas as pd

# Data for INPUT
input_data = {
    "TEST_ID": [100, 101, 102, 103, 104, 105],
    "MARKS": [55, 55, 60, 58, 40, 50]
}

# Creating the DataFrame
input_df = pd.DataFrame(input_data)

# Display the DataFrame
print(input_df)
```

```python
report=input_df.sort_values("TEST_ID")
report["IMP"]=report["MARKS"].diff(periods=1)
report
```

```python
import numpy as np
report[(pd.isna(report["IMP"])) | (report["IMP"]>0)][["TEST_ID","MARKS"]]
```

```python
report[(report["IMP"]> 0)][["TEST_ID","MARKS"]]
```
