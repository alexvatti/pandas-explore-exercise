```python
import pandas as pd

# Step 1: Create a list of tuples (employee, manager)
data = [
    ('Elon', None),
    ('Ira', 'Elon'),
    ('Bret', 'Elon'),
    ('Earl', 'Elon'),
    ('James', 'Ira'),
    ('Drew', 'Ira'),
    ('Mark', 'Bret'),
    ('Phil', 'Mark'),
    ('Jon', 'Mark'),
    ('Omid', 'Earl')
]

# Step 2: Create a DataFrame from the list
df = pd.DataFrame(data, columns=['employee', 'manager'])

# Display the DataFrame
print(df)
```

```python
grouped = df.groupby("manager")["employee"].apply(list).reset_index()
print(grouped)

l = grouped["manager"].to_list()
e = grouped["employee"].to_list()
print(l)
print(e)
```
