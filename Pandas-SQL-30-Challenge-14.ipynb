```python
import pandas as pd

# Step 1: Create a list of tuples representing the invoice data
data = [
    (330115, '2024-03-01'),
    (330120, '2024-03-01'),
    (330121, '2024-03-01'),
    (330122, '2024-03-02'),
    (330125, '2024-03-02')
]

# Step 2: Create a DataFrame from the list
df_invoice = pd.DataFrame(data, columns=['serial_no', 'invoice_date'])

# Step 3: Convert the 'invoice_date' column to datetime
df_invoice['invoice_date'] = pd.to_datetime(df_invoice['invoice_date'],)

# Display the DataFrame
print(df_invoice)
```

```python
df_invoice.sort_values("serial_no",inplace=True)
df_invoice
```

```python
df_invoice["diff"] = df_invoice["serial_no"].diff(periods=1)-1
```

```python
result = []
# Iterate over the DataFrame rows
for index, row in df_invoice.iterrows():
    if pd.isna(row['diff']) or row['diff'] == 0:
        pass
    else:
        num = row['diff']
        for i in range(1, int(num) + 1):
            result.append(row['serial_no'] - i)

# Display the result list
result=sorted(result)
 
pd.DataFrame(result,columns=["MISSING_SERIAL_NO"])
```
