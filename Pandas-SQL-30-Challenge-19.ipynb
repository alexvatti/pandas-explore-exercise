```python
import pandas as pd

df = pd.read_excel("EX-19.xlsx")
df
```

```python
df["ACTUAL_DELIVERY"].isnull().sum()
```

```python


# Convert time columns to datetime
df['ORDER_TIME'] = pd.to_datetime(df['ORDER_TIME'], format='%d-%m-%Y %H:%M')
df['EXPECTED_DELIVERY'] = pd.to_datetime(df['EXPECTED_DELIVERY'], format='%d-%m-%Y %H:%M')
df['ACTUAL_DELIVERY'] = pd.to_datetime(df['ACTUAL_DELIVERY'], format='%d-%m-%Y %H:%M')

# Check if the delivery was delayed (delayed if actual delivery is more than 30 minutes later than expected)
df['is_delayed'] = (df['ACTUAL_DELIVERY'] - df['EXPECTED_DELIVERY']) > pd.Timedelta(minutes=30)

# Extract the month and year for grouping
df['MONTH'] = df['ORDER_TIME'].dt.strftime('%Y-%m')

# Group by month and calculate delayed orders and free pizzas
monthly_stats = df.groupby('MONTH').agg(
    total_orders=('ORDER_ID', 'count'),
    delayed_orders=('is_delayed', 'sum'),
    free_pizzas=('NO_OF_PIZZAS', lambda x: x[df['is_delayed']].sum())
).reset_index()

# Calculate percentage of delayed orders
monthly_stats['delayed_percentage'] = (monthly_stats['delayed_orders'] / monthly_stats['total_orders']) * 100

# Sort by month
monthly_stats = monthly_stats.sort_values(by='MONTH')

# Display the result
print(monthly_stats)
```
