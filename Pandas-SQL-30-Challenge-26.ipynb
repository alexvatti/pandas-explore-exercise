```python
# Create a list of dictionaries based on the provided data
data = [
    {"TOKEN_NUM": 1, "CUSTOMER": "Maryam"},
    {"TOKEN_NUM": 2, "CUSTOMER": "Rocky"},
    {"TOKEN_NUM": 3, "CUSTOMER": "John"},
    {"TOKEN_NUM": 3, "CUSTOMER": "John"},
    {"TOKEN_NUM": 2, "CUSTOMER": "Arya"},
    {"TOKEN_NUM": 1, "CUSTOMER": "Pascal"},
    {"TOKEN_NUM": 9, "CUSTOMER": "Kate"},
    {"TOKEN_NUM": 9, "CUSTOMER": "Ibrahim"},
    {"TOKEN_NUM": 8, "CUSTOMER": "Lilly"},
    {"TOKEN_NUM": 8, "CUSTOMER": "Lilly"},
    {"TOKEN_NUM": 5, "CUSTOMER": "Shane"}
]
import pandas as pd

# Convert the list to a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
df.drop_duplicates(inplace=True)
result=df.groupby("TOKEN_NUM").agg(
    CUSTOMER_COUNT = ('CUSTOMER','count')
).reset_index()
result = result[result["CUSTOMER_COUNT"]==1].sort_values("CUSTOMER_COUNT").head(1)
result[["TOKEN_NUM"]]
```
