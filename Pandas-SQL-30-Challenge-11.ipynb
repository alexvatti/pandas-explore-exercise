```python
import pandas as pd

# List of tuples representing hotel ratings data
hotel_ratings_data = [
    ('Radisson Blu', 2020, 4.8),
    ('Radisson Blu', 2021, 3.5),
    ('Radisson Blu', 2022, 3.2),
    ('Radisson Blu', 2023, 3.8),
    ('InterContinental', 2020, 4.2),
    ('InterContinental', 2021, 4.5),
    ('InterContinental', 2022, 1.5),
    ('InterContinental', 2023, 3.8)
]

# Create a DataFrame from the list of tuples
df = pd.DataFrame(hotel_ratings_data, columns=['Hotel', 'Year', 'Rating'])

# Print the DataFrame
print(df)
```

```python
df['Average_Rating'] = df.groupby('Hotel')['Rating'].transform('mean')
df['std'] = df.groupby('Hotel')['Rating'].transform('std')
df["upper-limit"]=df["Average_Rating"]+df["std"]
df["lower-limit"]=df["Average_Rating"]-df["std"]
df[(df["Rating"]<df["upper-limit"]) &  (df["Rating"] > df["lower-limit"])][["Hotel","Year","Rating"]]
```
