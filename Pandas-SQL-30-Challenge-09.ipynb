```python
import pandas as pd

# Define the data as a list of tuples
data = [
    (1, '2024-02-18', 101),
    (1, '2024-02-18', 102),
    (1, '2024-02-19', 101),
    (1, '2024-02-19', 103),
    (2, '2024-02-18', 104),
    (2, '2024-02-18', 105),
    (2, '2024-02-19', 101),
    (2, '2024-02-19', 106)
]

# Create a DataFrame from the list
df = pd.DataFrame(data, columns=['customer_id', 'dates', 'product_id'])

# Print the DataFrame
print(df)
```

```python
df["dates"]=pd.to_datetime(df["dates"])
df
```

```python
merged_df = df.groupby(['customer_id', 'dates'])['product_id'].apply(list).reset_index()

# Print the merged DataFrame
print(merged_df)
```

```python
pd.concat([merged_df,df]).reset_index().sort_values("dates")
```
