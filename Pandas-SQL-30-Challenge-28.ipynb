```python
import pandas as pd

# Create a list of dictionaries based on the provided values
data = [
    {"ID": 1, "ITEMS": "22,122,1022"},
    {"ID": 2, "ITEMS": ",6,0,9999"},
    {"ID": 3, "ITEMS": "100,2000,2"},
    {"ID": 4, "ITEMS": "4,44,444,4444"}
]

# Convert the list into a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
def cal_str_length(txt):
    l = []
    for word in txt.split(","):
        l.append(len(word))
    return ",".join(map(str,l))
df["LENGTHS"]=df["ITEMS"].apply(cal_str_length)
df[["ID","LENGTHS"]]
```
